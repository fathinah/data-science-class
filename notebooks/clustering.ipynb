Question: If A's organization is keilmuan, nilai ujian tulis 60, nilai wawancara 62, and komunikasi is tinggi will A do an exchange?
> p(ya) = 9/14
<br> p(tidak) = 5/14
<br> asked: p(Ya|o=keilmuan, ut=60, w=62) vs p(Tidak|o=keilmuan, ut=60, w=62)
<br> p(k|ya).p(ut|ya).(w|ya).p(ya) vs p(k|tdk).p(ut|tdk).(w|tdk).p(tdk)

```python
ut_ya=[83,70,68,64,69,75,75,72,81]
w_ya=[86,96,80,65,70,80,70,90,75]
ut_tdk=[71,72,65,80,85]
w_tdk=[91,95,70,85,90]
```

```python
import math
import numpy as np
```

```python
mutya=np.mean(ut_ya)
muttdk=np.mean(ut_tdk)
# stdutya= np.std(ut_ya)
# stduttdk=np.std(ut_tdk)
mwya=np.mean(w_ya)
mwtdk=np.mean(w_tdk)
```

```python
#standard dev formula with divisor n-1
def stddev(lst,mean):
  tot=0
  for i in lst:
    j = i-mean
    j = j*j
    tot+=j
  return (tot/(len(lst)-1))**0.5
```

## Calculate prior likelihood of numeric attribute

```python
cons = (math.pi*2)**0.5
def dens_func(x,mean, std):
  ut_ya = math.pow(math.e,(-1*((x-mean)**2/(2*(std**2)))))
  temp = 1/(cons*std)
  return(temp*ut_ya)
```

```python
putya = (dens_func(60, mutya, stddev(ut_ya,mutya)))
puttdk= (dens_func(60, muttdk, stddev(ut_tdk,muttdk)))

pwya= (dens_func(62, mwya, stddev(w_ya,mwya)))
pwtdk= (dens_func(62, mwtdk, stddev(w_tdk,mwtdk)))
```

## Calculate prior likelihood of categorical attribute

```python
pya = 9/14
ptdk = 1-pya
poya=4/9
potdk=1/(5+3) #Plus laplace estimator
pkomya=6/9
pkomtdk=2/5
```

## Calculate posterior

```python
yagivenx = poya*pkomya*putya*pwya*pya
nogivenx = potdk*pkomtdk*puttdk*pwtdk*ptdk
```

```python
yagivenx
```

```python
nogivenx
```

> yes>no => YES, this person goes exchange
